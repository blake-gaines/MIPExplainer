```python
import pandas as pd
import numpy as np
import pickle
from datasets import get_dataset
import matplotlib.pyplot as plt
import plotly.express as px
from PIL import Image
```

```python
replace_d = {
    "runtime": "Runtime (s)",
    "consistency": "Average Edit Distance",
    "avg_consistency": "Average Edit Distance",
    "num_nodes": "\\# Nodes",
    "max_class": "Class",
    # "Is_Acyclic_Ones": "Is_Acyclic",
    # "Shapes_Ones": "Shapes",
    "method": "Method",
    "dataset_name": "Dataset",
    "_": "\\_",
    "NaN": "N/A",
    # "\\\\\n\\cline{1-6}": "\\vspace{0.2cm} \\\\\n\\cline{1-6}",
    "\\cline{1-6}\n\\bottomrule": "\\bottomrule",
}

def replace_all(text):
    for i, j in replace_d.items():
        text = text.replace(i, j)
    return text

methods = ["MIPExplainer", "GNNInterpreter", "XGNN", "PAGE"]
```

```python
def combine_images_horizontally(image_list, separator_width=10):
    # Ensure there are images to combine
    if not image_list:
        raise ValueError("The image list is empty")

    # Calculate total width and maximum height of the final image
    total_width = sum(image.width for image in image_list) + separator_width * (len(image_list) - 1)
    max_height = max(image.height for image in image_list)

    # Create a new image with a transparent background
    combined_image = Image.new('RGBA', (total_width, max_height), (0, 0, 0, 0))

    # Paste images into the combined image
    x_offset = 0
    for image in image_list:
        combined_image.paste(image, (x_offset, 0), image)
        x_offset += image.width + separator_width

    return combined_image
```

```python
dfs = []
datasets = {}
for dataset_name in ["MUTAG", "Is_Acyclic_Ones", "Shapes_Ones"]:
    dfs.append(pickle.load(open(f"results/all_{dataset_name}.pkl", "rb")))
    datasets[dataset_name] = get_dataset(dataset_name)
df = pd.concat(dfs).dropna(axis=1, how="any").sort_index()
df
```

```python
dataset_name = "Shapes_Ones"
```

```python
df = df.loc[dataset_name,:,:,:]
df
```

```python
df = df.groupby(df.index.names, group_keys=False).apply(
        lambda x: x.sample(3, random_state=0)
    )
```

```python
combined_images = df.groupby(df.index.names, group_keys=False).apply(
        lambda x: combine_images_horizontally(x["Image"].tolist())
    )
df = pd.DataFrame()
df["Image"] = combined_images
```

```python
## Create a latex table with images
df["i"] = list(range(len(df)))
df["title"] = df["i"].map(lambda x: ""+f"\\includegraphics[width=0.2\\textwidth]{{exps/{dataset_name}_exp_{x}.png}}"+"")
# df["title"] = df["i"].map(lambda x: f"\\includegraphics[width=0.2\\textwidth]{{exps/{dataset_name}_exp_{x}.png}}")
df["path"] = df["i"].map(lambda x: f'results/exps/{dataset_name}_exp_{x}.png')
```
